```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"../data/sampled_data.csv")
```

```python
df.head()
```

```python
# 1. BASIC LOOK AT DATA

print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
# 1. Drop unnecessary columns
df = df.drop(columns=['Unnamed: 0'])

# 2. Drop 1 missing row in question2
df = df.dropna(subset=['question2'])

print("✅ Clean shape:", df.shape)  
print(df.isnull().sum())         
```

```python
# 2. HOW MANY DUPLICATES VS NOT DUPLICATES
print(df['is_duplicate'].value_counts())

df['is_duplicate'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Duplicate vs Not Duplicate')
plt.xlabel('is_duplicate')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Duplicate', 'Duplicate'], rotation=0)
plt.show()
```

```python
# Save feature engineered data
df.to_csv('../data/cleaned_data.csv', index=False)
print("✅ Saved!", df.shape)
```
